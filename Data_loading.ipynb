Let's name the data frame as __'survey_raw_df'__ to indicate that this is unprocessed data and can be cleaned, filtered and modified to prepare a data frame that's ready for the analysis.

As we can see that the dataset contains over __60000 responses__ and around __100 questions__. Each respondent has been assigned a unique ramdomized respondent ID.

Let's review the list of columns in the data frame.

```python
list(survey_raw_df.columns)
```

Here, the column names are given as the short codes of questions.

We can refer to __schema file__ for the full question text.
Loading the schema CSV using pandas...

```python
schema_raw_df = pd.read_csv('stack-overflow-developer-survey-2024/survey_results_schema.csv')
```

```python
schema_raw_df
```

```python
schema_raw_df.qname
```

Now, we can see those short codes in __'qname' column and the questions in 'question' column. We can remove other useless columns and make 'qname' as index, so we can load dataset as pandas series with 'qname' as the index and the 'question' as the value.

```python
schema_raw_ndf = schema_raw_df.drop(columns=['qid', 'force_resp', 'type', 'selector'])
```

```python
schema_raw_ndf.set_index('qname', inplace=True)
```

```python
schema_raw = schema_raw_ndf.question    # gives the series of the dataset
```

Now, we can use the __'schema_raw'__ to retrieve any question text for any column in __'survey_raw_df'__

```python
schema_raw.AINext    # ex.
```
